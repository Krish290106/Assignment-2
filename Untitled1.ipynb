```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from statsmodels.stats.diagnostic import linear_rainbow
import statsmodels.api as sm
```

```python
 data = pd.read_csv("/content/tips (data for regression problem).csv")
data.head()
```

```python
sns.pairplot(data)
plt.show()
```

```python
num_df=data.select_dtypes(include=[np.number])
plt.figure(figsize=(8, 6))
sns.heatmap(num_df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
features = data[['total_bill']]
target = data['tip']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
predictions_lr = linear_reg.predict(X_test)
r2_lr = r2_score(y_test, predictions_lr)
mae_lr = mean_absolute_error(y_test, predictions_lr)
mse_lr = mean_squared_error(y_test, predictions_lr)

print("Linear Regression")
print("R Score:", r2_lr)
print("MAE:", mae_lr)
print("MSE:", mse_lr)
print("\n")
```

```python
ridge_reg = Ridge(alpha=1.0)
ridge_reg.fit(X_train, y_train)
predictions_ridge = ridge_reg.predict(X_test)
r2_ridge = r2_score(y_test, predictions_ridge)
mae_ridge = mean_absolute_error(y_test, predictions_ridge)
mse_ridge = mean_squared_error(y_test, predictions_ridge)

print("Ridge Regression")
print("R Score:", r2_ridge)
print("MAE:", mae_ridge)
print("MSE:", mse_ridge)
print("\n")
```

```python
lasso_reg = Lasso(alpha=0.1)
lasso_reg.fit(X_train, y_train)
predictions_lasso = lasso_reg.predict(X_test)
r2_lasso = r2_score(y_test, predictions_lasso)
mae_lasso = mean_absolute_error(y_test, predictions_lasso)
mse_lasso = mean_squared_error(y_test, predictions_lasso)

print("Lasso Regression")
print("R Score:", r2_lasso)
print("MAE:", mae_lasso)
print("MSE:", mse_lasso)
print("\n")
```

```python
tree_reg = DecisionTreeRegressor()
tree_reg.fit(X_train, y_train)
predictions_tree = tree_reg.predict(X_test)
r2_tree = r2_score(y_test, predictions_tree)
mae_tree = mean_absolute_error(y_test, predictions_tree)
mse_tree = mean_squared_error(y_test, predictions_tree)

print("Decision Tree")
print("R Score:", r2_tree)
print("MAE:", mae_tree)
print("MSE:", mse_tree)
print("\n")
```

```python
forest_reg = RandomForestRegressor(n_estimators=100)
forest_reg.fit(X_train, y_train)
predictions_forest = forest_reg.predict(X_test)
r2_forest = r2_score(y_test, predictions_forest)
mae_forest = mean_absolute_error(y_test, predictions_forest)
mse_forest = mean_squared_error(y_test, predictions_forest)

print("Random Forest")
print("R Score:", r2_forest)
print("MAE:", mae_forest)
print("MSE:", mse_forest)
print("\n")
```

```python
svr_reg = SVR(kernel='rbf')
svr_reg.fit(X_train, y_train)
predictions_svr = svr_reg.predict(X_test)
r2_svr = r2_score(y_test, predictions_svr)
mae_svr = mean_absolute_error(y_test, predictions_svr)
mse_svr = mean_squared_error(y_test, predictions_svr)

print("SVR")
print("R Score:", r2_svr)
print("MAE:", mae_svr)
print("MSE:", mse_svr)
print("\n")
```

```python
knn_reg = KNeighborsRegressor(n_neighbors=5)
knn_reg.fit(X_train, y_train)
predictions_knn = knn_reg.predict(X_test)
r2_knn = r2_score(y_test, predictions_knn)
mae_knn = mean_absolute_error(y_test, predictions_knn)
mse_knn = mean_squared_error(y_test, predictions_knn)

print("KNN Regression")
print("R Score:", r2_knn)
print("MAE:", mae_knn)
print("MSE:", mse_knn)
print("\n")
```

##CONCLUSION
Based on the performance metrics (R Score, MAE, and MSE) for each regression model, here are the key takeaways:

1. **Lasso Regression, Ridge Regression, and Linear Regression**:
   - These three models performed similarly with R Scores around 0.540.55, MAEs close to 0.62, and MSEs near 0.57.
   - This suggests that linear models are providing reasonable performance, and Lasso and Ridge regularization don't show significant improvements over standard Linear Regression for this dataset.

2. **Support Vector Regression (SVR)**:
   - SVR has an R Score of 0.51, slightly lower than the linear models, but it has the lowest MAE (0.61) and a competitive MSE (0.61).
   - This implies that SVR is performing comparably well in terms of average error (MAE) but slightly lags in capturing overall variance (R Score).

3. **K-Nearest Neighbors (KNN) Regression**:
   - KNN has an R Score of 0.45, which is lower than the linear models and SVR.
   - Its MAE (0.68) and MSE (0.68) are also higher, indicating that it is less effective on this dataset compared to other models.

4. **Random Forest**:
   - Random Forest has an R Score of 0.24, which is relatively low, indicating limited accuracy in capturing the variance in the data.
   - Its MAE (0.78) and MSE (0.95) are also high compared to other models, suggesting that it may be overfitting on this dataset or that the data may not benefit from a more complex ensemble approach.

5. **Decision Tree**:
   - The Decision Tree model has the worst performance with a negative R Score (-0.07), indicating that it is performing worse than simply predicting the mean value of the target variable.
   - It also has the highest MAE (0.95) and MSE (1.34), showing that it is poorly suited for this dataset in its current configuration.

### Conclusion:
- **Best Models**: The linear models (Lasso, Ridge, and Linear Regression) provide the best balance of R Score, MAE, and MSE for this dataset, suggesting that a simple linear approach is effective.
- **SVR** also performs reasonably well and could be considered if a slightly lower MAE is desired, though it sacrifices some explanatory power (R Score).
- **Underperforming Models**: Decision Tree and Random Forest models show relatively poor performance, possibly due to overfitting or a lack of complex relationships in the data that would benefit from these approaches.
